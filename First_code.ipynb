```python
students = []

while True: # Capital T!
    name = input("Enter name: ")
    age = input("Enter age: ")
    grade = input("Enter grade: ")

    student_info = {
        "name": name,
        "age": age,
        "grade": grade
    }

    # (Ghost brace deleted)

    students.append(student_info)
    print(f"Added {name} to the list!") # Added 'f' and fixed quotes

    cont = input("Add another student? (yes/no): ")
    if cont.lower() == "no":
        break

print("\nFinal Student List:")
print(students)



```
