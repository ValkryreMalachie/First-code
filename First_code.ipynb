```python
def join_with_and(items):
    ## This function takes a list of items and formats them into a grammatically correct string.
    ## For example, ['A', 'B', 'C'] becomes 'A, B, and C'.
    if not items:
        return ""
    if len(items) == 1:
        return str(items[0])
    if len(items) == 2:
        return f"{items[0]} and {items[1]}"
    return f"{', '.join(items[:-1])}, and {items[-1]}"

students = []

while True: # Capital T!
    name = input("Enter name: ")
    age = input("Enter age: ")
    grade = input("Enter grade: ")

    student_info = {
        "name": name,
        "age": age,
        "grade": grade
    }

    # (Ghost brace deleted)

    students.append(student_info)
    print(f"Added {name} to the list!") # Added 'f' and fixed quotes

    cont = input("Add another student? (y/n): ")
    if cont.lower() == "n":
        break

print("\nFinal Student List:")

grade_count = {}

for s in students:
  grade = s['grade']
  if grade in grade_count:
    grade_count[grade] += 1 # Corrected to increment the count
  else:
    grade_count[grade] = 1 # Corrected else block indentation and assignment

grade_summaries = []

for grade, count in grade_count.items(): # Corrected 'grade_counts' to 'grade_count'
    s_suffix = "" if count == 1 else "s"
    grade_summaries.append(f"{count} student{s_suffix} in grade {grade}")

print(f"There are {join_with_and(grade_summaries)}.")
## Using list comprehension to extract all student names into a new list
all_names = [s['name'] for s in students]
## Using list comprehension to extract all student ages into a new list
all_ages = [s['age'] for s in students]
## Using list comprehension to extract all student grades into a new list
all_grades = [s['grade'] for s in students]
print(f"Welcome students {join_with_and(all_names)}, you are aged {join_with_and(all_ages)}, and are in grades {join_with_and(all_grades)}.")
```

```python
print(f"All Ages: {', '.join(all_ages)}")
print(f"All Grades: {', '.join(all_grades)}")
```
